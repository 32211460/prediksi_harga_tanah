```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Data kota-kota
kota_kabupaten = ['Kabupaten Bandung', 'Kabupaten Bandung Barat', 'Kabupaten Bekasi', 
                  'Kabupaten Bogor', 'Kabupaten Ciamis', 'Kabupaten Cianjur',
                  'Kabupaten Cirebon', 'Kabupaten Garut', 'Kabupaten Indramayu', 'Kabupaten Karawang',
                  'Kabupaten Kuningan', 'Kabupaten Majalengka', 'Kabupaten Pangandaran', 'Kabupaten Purwakarta', 
                  'Kabupaten Subang', 'Kabupaten Sukabumi', 'Kabupaten Sumedang', 'Kabupaten Tasikmalaya',  
                  'Kota Bandung', 'Kota Banjar', 'Kota Bekasi',
                  'Kota Bogor', 'Kota Cimahi', 'Kota Cirebon', 'Kota Depok', 'Kota Sukabumi', 'Kota Depok', 'Kota Bekasi', 'Kota Tasikmalaya']

# Data luas tanah
luas_tanah = np.array([80, 49, 200, 89, 80, 95, 39, 100, 93, 90, 70, 80, 72, 60, 50, 60, 180, 32, 90, 72, 110, 100, 30, 50, 40, 90, 75, 66, 78,])

# Data tahun tanah
tahun_tanah = np.array([2023]* len(kota_kabupaten))

# Data atribut tambahan
tingkat_pembangunan = np.array([5, 3, 4, 2, 4, 1, 5, 3, 4, 5, 4, 2, 4, 1, 5, 3, 4, 5, 3, 4, 110, 100, 30, 50, 40, 90, 75, 66, 78,])

akses_transportasi = np.array([4, 2, 3, 1, 3, 1, 5, 4, 5, 3, 5, 3, 2, 4, 5, 1, 2, 4, 3, 2, 2, 4, 5, 4, 3, 2, 4, 5, 1,])

fasilitas_umum = np.array([3, 2, 4, 1, 4, 1, 5, 3, 1, 3, 4, 2, 4, 1, 5, 3, 4, 5, 3, 4,  5, 4, 5, 5, 3, 2, 4, 5, 4, ])

kondisi_geografis = np.array([2, 3, 2, 1, 2, 3, 1, 2, 5, 3, 4, 1, 4, 1, 5, 3, 1, 3, 4, 5,  3, 4, 3, 5, 5, 1, 3, 2, 1, ])

potensi_investasi = np.array([4, 1, 3, 1, 2, 1, 5, 3, 3, 2, 4, 1, 4, 1, 5, 3, 1, 3, 4, 5,  3, 4, 3, 5, 5, 1, 3, 2, 1, ])
demografi_penduduk = np.array([1000000, 500000, 200000, 300000, 800000, 150000, 300000, 600000, 700000, 900000, 1200000, 500000, 300000,800000, 1300000, 1000000, 300000, 800000,
                              800000, 150000, 800000, 150000, 300000, 600000, 700000, 900000, 1200000, 500000, 300000, ])
kriminalitas = np.array([20, 5, 15, 3, 10, 2, 25, 8, 5, 9, 12, 3, 9, 7, 12, 21, 1, 3, 8, 11, 12, 3, 9, 7, 12, 21, 1, 3, 8 ]) 



# Membuat DataFrame 
df = pd.DataFrame({
    'Kota/Kabupaten': kota_kabupaten,
    'Luas Tanah': luas_tanah,
    'Tahun Tanah': tahun_tanah,
    'Tingkat Pembangunan': tingkat_pembangunan,
    'Akses Transportasi': akses_transportasi,
    'Fasilitas Umum': fasilitas_umum,
    'Kondisi Geografis': kondisi_geografis,
    'Potensi Investasi': potensi_investasi,
    'Demografi Penduduk': demografi_penduduk,
    'Kriminalitas': kriminalitas
})

# Menghitung harga tanah 
df['Harga Tanah (Juta)'] = 0.005 * df['Luas Tanah'] + 0.002 * df['Tingkat Pembangunan'] + \
                             0.0015 * df['Akses Transportasi'] + 0.001 * df['Fasilitas Umum'] + \
                             0.0005 * df['Kondisi Geografis'] + 0.0025 * df['Potensi Investasi'] + \
                             0.0002 * df['Demografi Penduduk'] - 0.0005 * df['Kriminalitas'] + \
                             np.random.normal(scale=0.1, size=len(df))

# Menyimpan DataFrame ke dalam file Excel
df.to_excel('data_prediksi_harga_tanah_JABAR.xlsx', index=False)
df
```
